Realice una gráfica de barras sobre la polbación de cada 
ciudad, ordenela de menor a mayo y responda: 
 
a. ¿Cuál es la segunda ciudad más poblada? 
b. ¿Qué posición ocupa Granada en el ranking de 
las ciudades más pobladas? 

```python
data_ordenada = data.sort_values('Poblacion')

plt.figure(figsize=(10, 8))
plt.barh(data_ordenada['Municipio'], data_ordenada['Poblacion'], color='skyblue')
plt.xlabel('Población')
plt.ylabel('Municipio')
plt.title('Población de los Municipios Ordenados de Menor a Mayor')
plt.grid(axis='x')

plt.barh(data_ordenada['Municipio'].tail(10), data_ordenada['Poblacion'].tail(10), color='skyblue')

plt.show()

segunda_ciudad_mas_poblada = data_ordenada['Municipio'].iloc[-2]


posicion_granada = data_ordenada[data_ordenada['Municipio'] == 'Granada'].index[0]
ranking_posicion_granada = data_ordenada.index.tolist().index(posicion_granada) + 1

print("La segunda ciudad más poblada es:", segunda_ciudad_mas_poblada)
print("Granada ocupa la posición", ranking_posicion_granada, "en el ranking de las ciudades más pobladas.")
```

¿Cuántos municipios de Extremadura tienen más de 5000 
habitantes?. 

```python

municipios_extremadura_badajoz = data[(data['Provincia'] == "Badajoz") & (data['Poblacion'] > 5000)]
municipios_extremadura_caceres = data[(data['Provincia'] == "Cáceres") & (data['Poblacion'] > 5000)]

cantidad_municipios_extremadura_badajoz = municipios_extremadura_badajoz.shape[0]
cantidad_municipios_extremadura_caceres = municipios_extremadura_caceres.shape[0]

total_municipios_extremadura_5000 = cantidad_municipios_extremadura_badajoz + cantidad_municipios_extremadura_caceres

print("Número de municipios en Extremadura con más de 5,000 habitantes:", total_municipios_extremadura_5000)
```

¿Cuál es el municipio situado más al Norte? (Usar el 
valor de la coordenada "Y" que representa la latitud en 
grados). Proporcione también la provincia a la que 
pertenece y su población. 

```python

municipio_mas_al_norte = data.loc[data['Y'].idxmax()]

nombre_municipio_norte = municipio_mas_al_norte['Municipio']
provincia_municipio_norte = municipio_mas_al_norte['Provincia']
poblacion_municipio_norte = municipio_mas_al_norte['Poblacion']

print("Municipio más al norte:", nombre_municipio_norte)
print("Provincia:", provincia_municipio_norte)
print("Población:", poblacion_municipio_norte)
```

Obtenga la media, mediana, desviación estándar, valor 
máximo y valor mínimo de la población de los municipios 
de la provincia de Granada. 

```python

municipios_granada = data[data['Provincia'] == 'Granada']

media_poblacion_granada = municipios_granada['Poblacion'].mean()

mediana_poblacion_granada = municipios_granada['Poblacion'].median()

std_poblacion_granada = municipios_granada['Poblacion'].std()

max_poblacion_granada = municipios_granada['Poblacion'].max()

min_poblacion_granada = municipios_granada['Poblacion'].min()

print("Media de la población:", media_poblacion_granada)
print("Mediana de la población:", mediana_poblacion_granada)
print("Desviación estándar de la población:", std_poblacion_granada)
print("Valor máximo de la población:", max_poblacion_granada)
print("Valor mínimo de la población:", min_poblacion_granada)
```

 Realice un histograma con la población de los 
Municipios para cada una de las provincias. 

```python

provincias = data['Provincia'].unique()

num_provincias = len(provincias)
num_filas = int(num_provincias ** 0.5) + 1
num_columnas = (num_provincias // num_filas) + (num_provincias % num_filas > 0)

plt.figure(figsize=(num_columnas * 5, num_filas * 5))

for i, provincia in enumerate(provincias, 1):
    plt.subplot(num_filas, num_columnas, i)
    poblacion_municipios = data[data['Provincia'] == provincia]['Poblacion']
    plt.hist(poblacion_municipios, bins=20, color='skyblue', edgecolor='black')
    plt.title(provincia)
    plt.xlabel('Población')
    plt.ylabel('Número de Municipios')
    plt.tight_layout()

plt.show()
```

Seleccione al azar cincuenta municipios diferentes de

```python

municipios_aleatorios = data.sample(n=50, replace=False)

promedio_poblacion_aleatorios = municipios_aleatorios['Poblacion'].mean()

desviacion_std_poblacion_aleatorios = municipios_aleatorios['Poblacion'].std()

print("Promedio de la población de los 50 municipios seleccionados:", promedio_poblacion_aleatorios)
print("Desviación estándar de la población de los 50 municipios seleccionados:", desviacion_std_poblacion_aleatorios)
```

Determine los nombres de los Municipios más cercano y 
más lejano a Madrid. Para ello debe calcular la 
distancia en todos ellos y Madrid. No considere a Madrid 
en el análisis ya que la distancia sería cero. 

```python

data_sin_madrid = data[data['Municipio'] != 'Madrid']

madrid_x = data[data['Municipio'] == 'Madrid']['X'].values[0]
madrid_y = data[data['Municipio'] == 'Madrid']['Y'].values[0]

distancias = np.sqrt((data_sin_madrid['X'] - madrid_x)**2 + (data_sin_madrid['Y'] - madrid_y)**2)

data_sin_madrid['Distancia'] = distancias

municipio_mas_cercano = data_sin_madrid.loc[data_sin_madrid['Distancia'].idxmin()]
municipio_mas_lejano = data_sin_madrid.loc[data_sin_madrid['Distancia'].idxmax()]

nombre_municipio_cercano = municipio_mas_cercano['Municipio']
nombre_municipio_lejano = municipio_mas_lejano['Municipio']

(nombre_municipio_cercano, nombre_municipio_lejano)
```
